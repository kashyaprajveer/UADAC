```python
#Dependencies
import pandas as pd
import numpy as np
import tensorflow as tf
import itertools
import matplotlib.pyplot as plt
import time

from envs import OfflineEnv
from recommender import DRRAgent

import os
STATE_SIZE = 10
MAX_EPISODE_NUM = 20001

os.environ["CUDA_VISIBLE_DEVICES"]="1"
```

```python
movies_df=pd.read_csv('./movies.dat')
movies_df.head(2)
```

```python
movies_list=movies_df.values.tolist()
movies_list
```

```python
movies_id_to_movies = {movie[0]: movie[1:] for movie in movies_list}
```

```python
movies_id_to_movies
```

```python
movies_df['MovieID'] = movies_df['MovieID'].apply(pd.to_numeric)
```

```python
ratings_df=pd.read_csv('./ratings.dat')
ratings_df = ratings_df.applymap(int)
ratings_df.head(2)
```

```python
users_dict = np.load('./user_dict.npy', allow_pickle=True)
```

```python
users_history_lens = np.load('./users_histroy_len.npy')
```

```python
users_num = max(ratings_df["UserID"])+1  # Total users + 1
items_num = max(ratings_df["MovieID"])+1
```

```python
train_users_num = int(users_num * 0.8) # First 80% of users data
train_items_num = items_num            # total movies
train_users_dict = {k:users_dict.item().get(k) for k in range(1, train_users_num+1)} # user :(movieID, ratings) dictionary
train_users_history_lens = users_history_lens[:train_users_num]                # First 80% user's movies interaction data.

print('DONE!')
time.sleep(2)
```

```python
    env = OfflineEnv(train_users_dict, train_users_history_lens, movies_id_to_movies, STATE_SIZE)
    # train_users_dict - array({1: [(3186, 4), (1721, 4), (1270, 5), (1022, 5), (2340, 3), (1836, 5), (3408, 4), (1207, 4), (2804, 5), (260, 4), (720, 3), (1193, 5), (919, 4)]
    # train_users_history_lens - Total number of movies rated
    # movies_id_to_movies  {'1': ['Toy Story (1995)', "Animation|Children's|Comedy"],....
    # STATE_SIZE = 10
    recommender = DRRAgent(env, users_num, items_num, STATE_SIZE, use_wandb=False)
    recommender.actor.build_networks()
    recommender.actor2.build_networks()
    recommender.critic.build_networks()
    recommender.critic2.build_networks()
    recommender.train(MAX_EPISODE_NUM, load_model=False)
```
